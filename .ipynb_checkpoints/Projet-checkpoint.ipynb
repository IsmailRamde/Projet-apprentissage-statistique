# Importation des packages

```python
# standard imports
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import pandas as pd
import seaborn as sns
```

# Chargement des données

```python
X_clinical  = pd.read_csv("dataset/train-data_clinical.csv", sep = ";")

X_otu  = pd.read_csv("dataset/train-data_otu.csv", sep = ";", low_memory=False)
```

```python
X_clinical_test  = pd.read_csv("dataset/test-data_clinical.csv", sep = ";")
X_otu_test  = pd.read_csv("dataset/test-data_otu.csv", sep = ";", low_memory=False)
```

```python
y_label = pd.read_csv ('dataset/train-label.txt', header=None, index_col=False)
y_label.to_csv ('dataset/train-label2.csv', index=None)
y_label.columns = ["Status"]
```

# Pré-processing

```python
code = {'Healthy':0,
        'IBD':1}

for i in y_label.select_dtypes('object').columns:
    y_label[i] = y_label[i].map(code)

y_label.astype(int)
```

```python
def pre_processing(data):
    column_data = np.ravel(np.array(data.columns))
    for i in column_data:
        if data[i].dtypes=="object":
            data[i] = [x.replace(',','.') for x in data[i]]
    data = data.apply(pd.to_numeric)
    return data

X_otu = pre_processing(X_otu)

X_otu_test = pre_processing(X_otu_test)
```

```python
# Concatenat X
X_data = pd.concat([X_clinical,X_otu], axis=1)
```

# Lasso

```python


from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
parameters={'alpha':[0,0.1,0.2,0.3,0.5,0.5,0.6,0.7,0.8,0.9,1]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)

lasso_regressor.fit(X_data,y_label)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
X_data_test = pd.concat([X_clinical_test,X_otu_test], axis=1)
prediction_lasso=lasso_regressor.predict(X_data_test)
prediction_lasso
```

```python
import seaborn as sns

#sns.distplot(y_label-prediction_lasso)
lasso_regressor.score(X_data, y_label)
```
